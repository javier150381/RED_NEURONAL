### Ahora vavos a mostrarle al modelo una imagen que nunca ha visto para ver como predice si un paciente tiene o no neumonía de acuerdo a la radiografia de torax

#### Importamos las librerías necesaria

```python
from keras.models import load_model
import matplotlib.pyplot as plt
from skimage.transform import resize
from PIL import Image
import numpy as np
import matplotlib.image as mpimg
import tensorflow as tf 
```

#### Llamamos al modelo

```python

model = load_model('NEUMONIA.h5py')
```

#### Cargamos la imagen

```python
path="c:/prueba/neumo1.jpeg"
plt.imshow(mpimg.imread(path))
```

#### Definimos una función que nos permita mostrar el resultado depuès de la predicción

```python
def opcion(op):
    if op==0 :
        print("Normal")
    if op==1 :
        print("Neumonía")
    
```

#### Redimensionamos la imagen cargada para cumplir con los parámetros del modelo

```python
images=[]
filenames = [path,path]
```

```python
for filepath in filenames:
    image = plt.imread(filepath)
    image_resized = resize(image, (28, 28),preserve_range=True)
    images.append(image_resized)
```

```python
X = np.array(images, dtype=np.uint8) #convierto de lista a numpy
test_X = X.astype('float32')
test_X = test_X / 255.
```

#### Realizamos la predicción

```python
predicted_classes = model.predict(test_X)
prediccion = predicted_classes[0]
np.argmax(prediccion)
fg=int(np.argmax(prediccion))
print(fg)
```

#### Mostramos el resultado de la predicción

```python
opcion(fg)
```

#### La predicción es correcta ya que la imagen expuesta es de una persona que tiene neumonía
